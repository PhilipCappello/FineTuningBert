```python
import pandas as pd 
```

```python
cols = ["date", "tweet", "sentiment"]
df_new = pd.read_csv('CSVFiles/bitcointweets-new.csv')
df_new = df_new.iloc[:,[0,1,7]]
df_new.columns = cols 
df_new
```

```python
# convert all textual sentiment to numerical values 

for index, row in df_new.iterrows(): 
    if row['sentiment'] == "['neutral']":
        row['sentiment'] = 2 
    elif row['sentiment'] == "['positive']":
        row['sentiment'] = 1
    else:
        row['sentiment'] = 0
```

```python
# extract neutral tweets 
df_neutral = df_new.loc[df_new['sentiment']==2]
df_neutral
```

```python
df_old = pd.read_csv('CSVFiles/training-data-positive-and-negative-sentiment-english.csv')
```

```python
df_old = df_old.drop('Unnamed: 0',axis=1)
# df_old.rename(columns={"Date": "date", "text":"tweet","Sentiment": "sentiment"})
df_old.columns = ['date', 'tweet', 'sentiment']
```

```python
for index, row in df_old.iterrows(): 
    if row['sentiment'] == "Positive":
        row['sentiment'] = 1
    else:
        row['sentiment'] = 0
```

```python
df_old
```

```python
df_old_added_neutral = pd.concat([df_old, df_neutral], ignore_index=True)
df_old_added_neutral
```

```python
df_old_added_neutral.to_csv('CSVFiles/bitcointweets-old-added-neutral.csv')
```
